```python
import os
```

```python
from keras.preprocessing import image
import cv2
```

```python
categories =['with_mask','without_mask']
```

```python
data=[]
for category in categories:
    path=os.path.join('train',category)
#     print(path)
    label=categories.index(category)
    for file in os.listdir(path):
        img_path=os.path.join(path,file)
        img = cv2.imread(img_path)
        img =cv2.resize(img,(224,224))
#         print(img.shape)
        data.append([img,label])
    
```

```python
import random
```

```python
random.shuffle(data) #suffle the data
```

```python
x=[]
y=[]
for features,label in data:  #data is list
    x.append(features)
    y.append(label)
```

```python
import numpy as np
```

```python
x=np.array(x)
y=np.array(y)
```

```python
type(x),type(y)
```

```python
x.shape #( no of imges,height, width,no of channels(RGB) )
```

```python
x=x/255
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2)
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16()
```

```python
vgg.summary()
```

```python
#we remove last layer and applied over layer
```

```python
from keras import Sequential
model=Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary() #last layer is removed
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.summary() #last layer is removed
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
#4,097 trainable parameter 
#4096 bet fc2 and last dense layer
#bias of last dense layer
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
def detect_face_mask(img): #for given image return 1 if mask otherwise 0
#     y_pred = model.predict_classes(img.reshape(1,224,224,3))
    y_pred = (model.predict(img.reshape(1,224,224,3))>0.5).astype('int32')
    return y_pred[0][0]
```

```python
sample1 =cv2.imread('sample/1.jpg')
sample1=cv2.resize(sample1,(224,224))
```

```python
detect_face_mask(sample1)
```

```python
cap=cv2.VideoCapture(0)
```

```python
def draw_label(img,text,pos,bg_color):

    text_size =cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)
    
    end_x=pos[0]+text_size[0][0] + 2
    end_y=pos[1]+text_size[0][1] - 2    
    
    cv2.rectangle(img,pos,(end_x,end_y),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),cv2.LINE_AA)
```

```python
while True:
    ret,frame=cap.read()
    
    #the detect method
    img =cv2.resize(frame,(224,224))
    y_pred= detect_face_mask(img)
#     print(y_pread)
    if y_pred == 0: #if mask
        draw_label(frame,"Mask",(30,30),(0,225,0)) #green
    else:
        draw_label(frame,"No Mask",(30,30),(0,0,225)) #red
    
    cv2.imshow("window",frame)
    if cv2.waitKey(1) & 0XFF == ord('x'):
        break
cv2.destroyAllWindows()
```
